## Question 1

```python
import numpy as np
import pandas as pd
import random
import statistics as st
```

```python
nums = np.array([random.randint(0, 100) for r in range(0, 101)])
```

```python
nums
```

```python
def func1(nums):
    mean = np.mean(nums)
    median = np.median(nums)
    mode = st.mode(nums)
    std = np.std(nums)
    var = np.var(nums)
    return mean, median, mode, std, var
```

```python
print("The values for mean, median, mode, standard deviation and variance are as follows,")
print(func1(nums))
```

```python
# Finding quartiles, percentiles and IQR

q1 = np.quantile(nums, 0.25)
q3 = np.quantile(nums, 0.75)
iqr = q3 - q1

p25 = np.percentile(nums, 25)
p50 = np.percentile(nums, 50)
p75 = np.percentile(nums, 75)
p100 = np.percentile(nums, 100)
```

```python
print("The Q1 and Q3 values are ", q1, " and ", q3)
print("The IQR is ", iqr)
print("The 25th, 50th, 75th and 100th percentile values are ", p25, ", ", p50, ", ", p75, " and ", p100, " respectively")
```
